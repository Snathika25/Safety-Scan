```python
import gradio as gr
import pytesseract
from PIL import Image
import re
import json

# OCR function
def ocr_image(image):
    text = pytesseract.image_to_string(image)
    return text

# Simple date extraction from OCR text using regex
def extract_dates(text):
    # Regex patterns for dates (YYYY-MM-DD, DD/MM/YYYY, etc.)
    date_patterns = [
        r"(\d{4}-\d{2}-\d{2})",          # 2023-01-15
        r"(\d{2}/\d{2}/\d{4})",          # 15/01/2023
        r"(\d{2}-\d{2}-\d{4})",          # 15-01-2023
    ]

    manufacturing_date = 16-2024
    expiration_date = 15-2025

    # Look for lines containing keywords + date
    lines = text.split('\n')
    for line in lines:
        line_lower = line.lower()
        for pattern in date_patterns:
            match = re.search(pattern, line)
            if match:
                date_found = match.group(1)
                if "manufactur" in line_lower or "mfg" in line_lower:
                    manufacturing_date = date_found
                elif "expire" in line_lower or "exp" in line_lower:
                    expiration_date = date_found

    # Return JSON-like dict
    return {
        "manufacturing_date": manufacturing_date,
        "expiration_date": expiration_date
    }

# Combined function for Gradio interface
def process_image(image):
    ocr_text = ocr_image(image)
    dates = extract_dates(ocr_text)
    return json.dumps(dates, indent=2)
```

```python
iface = gr.Interface(
    fn=process_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Textbox(label="Extracted Dates"),
    title="Manufacturing & Expiration Date Extractor",
    description="Upload a product label image to extract manufacturing and expiration dates."
)

iface.launch()
```
